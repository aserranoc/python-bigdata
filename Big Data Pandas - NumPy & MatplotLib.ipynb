```python
#Series es muy similar al Json, codigo:valor
import pandas as pd
```

```python
series_lista = pd.Series([90,87,80,67],index=['FC Barcelona','Real Madrid', 'Atletico Madrid', 'Valencia'])
series_lista
```

```python
df_series_lista = pd.DataFrame(series_lista)
df_series_lista
```

```python
import numpy as np

#creando un array lineal random con numpy y convertirlo en DataFrame
array_uno = np.random.rand(5)
pd.DataFrame(array_uno)
```

```python
#creando una lista normal con python
lista_python = ['a','abc',12.2, False, True]
df_lista_python = pd.DataFrame(lista_python)
df_lista_python
```

```python
#creando un diccionario de datos con python
diccionario_python = {'ID001':'Cliente 1',
                     'ID002':'Cliente 2',
                     'ID003':'Cliente 3',
                     'ID004':'Cliente 4',
                     'ID005':'Cliente 5',
                     'ID006':'Cliente 6',
                     'ID007':'Cliente 7'}
#lo convierto a DataFrame con panda indicando que es de tipo dict con el from_dict
df_diccionario_python = pd.DataFrame.from_dict(diccionario_python, orient='index')
df_diccionario_python
```
